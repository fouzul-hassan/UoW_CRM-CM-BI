```python
#import pandas
import pandas as pd
# load dataset
data = pd.read_csv('/content/slr.csv')
data.head()
```

```python
X = data[['chirps per second']]
y = data['temperature (F)']
```

```python
#plot a chart
import matplotlib.pyplot as plt
plt.scatter(X, y)
plt.xlabel('chirps per second')
plt.ylabel('temperature (F)')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
slr_model = LinearRegression()
slr_model.fit(X_train, y_train)
```

```python
slope = slr_model.coef_
y_intercept = slr_model.intercept_
```

```python
y_intercept
```

```python
slope
```

```python
#plot the chart with the fitline
#figure size
plt.figure(figsize=(10, 6))
plt.scatter(X, y)
plt.xlabel('chirps per second')
plt.ylabel('temperature (F)')
plt.plot(X, slr_model.predict(X), color='red')
plt.show()
```

```python
y_pred_test = slr_model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_test = r2_score(y_test, y_pred_test)
mse_test = mean_squared_error(y_test, y_pred_test)
```
